```python
# Workable Code
# 1  Copy all files in a folder to another
import os
import shutil

# folder_path = "D:\\DATASETS\\UNBC\\Frame_Labels\\PSPI\\042-ll042\\ll042t1aaaff"
folder_path = "D:\\DATASETS\\UNBC\\Images"

for foldername in os.listdir(folder_path):
    # print("foldername: ",foldername)
    frame_labels = os.path.join(folder_path,foldername)
    for frame_label_persons in os.listdir(frame_labels):
        persons = os.path.join(frame_labels,frame_label_persons)
        # print(persons)
        for person in os.listdir(persons):
            txt_path = os.path.join(persons,person)
            
            src_folder = txt_path
            dst_folder = os.path.join('D:\\UNBC_DATASET\\IMAGES', person)
            # print(dst_folder)
            shutil.copy2(src_folder, dst_folder)
```
